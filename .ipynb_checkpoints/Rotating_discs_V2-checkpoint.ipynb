```python
from math import pi,sqrt
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Image
display(Image(filename="rd_formulae.png"));display(Image(filename="press_fit_pressure.png"));display(Image(filename="rd_strain.png"));display(Image(filename="rd_outer_inner_delta_r.png"))
```

```python
Ro = 0.32 / 2 #m
Ri = 0.03 #m

rho = 7830 #kg/m3
v = .288 

pressure = 85e6 #MPa
w = 6500 * 2*pi / 60 #rpm
E = 210e9 #GPa
```

# Bush Interference Fit

```python
delta_r = (2*Ri*pressure / E ) * 1 / (1 - Ri**2 / Ro**2)
print(f"radial interference {delta_r*1e6:.2f} microns")
delta_r_shaft = (Ri*-pressure / E) * (1 - v) 
print(f"shaft radial change {delta_r_shaft*1e6:.2f} microns")
delta_r_wheel = (Ri*pressure / E) * ((Ro**2 + Ri**2)/(Ro**2 - Ri**2) + v) 
print(f"flywheel radial change {delta_r_wheel*1e6:.2f} microns")
```

# Expansion With Speed

```python
B_shaft = 0
A_shaft = B_shaft / Ri**2 + (1/8)*(3+v)*rho*(w**2)*(Ri**2)
print(f"A = {A_shaft:.2f}, B = {B_shaft:.2f}")

r = np.linspace(Ri,Ro,100)
stress_radial = np.array(A_shaft - B_shaft / r**2 - (1/8)*(3+v)*rho*(w**2)*r**2)
stress_hoop = np.array(A_shaft + B_shaft / r**2 - (1/8)*(1+3*v)*rho*(w**2)*r**2)
stress_axial = np.zeros(len(r))
stress = np.array([stress_radial, stress_hoop, stress_axial])
vm_stress = (1/sqrt(2)) * np.sqrt( (stress[0]-stress[1])**2 + (stress[1]-stress[2])**2 + (stress[2]-stress[0])**2 )

delta_r_shaft = (Ri / E) * (stress_hoop[0] - v * (stress_radial[0] + stress_axial[0]))
print(f"radial change {delta_r_shaft*1e6:.2f} microns")
```

```python
B_flywheel = (- (1/8)*(3+v)*rho*(w**2)*(Ro**2-Ri**2)) / ( (1/Ro**2) - (1/Ri**2) )
A_flywheel = B_flywheel / Ro**2 + (1/8)*(3+v)*rho*(w**2)*(Ro**2)
print(f"A = {A_flywheel:.2f}, B = {B_flywheel:.2f}")

r = np.linspace(Ri,Ro,100)
stress_radial = np.array(A_flywheel - B_flywheel / r**2 - (1/8)*(3+v)*rho*(w**2)*r**2)
stress_hoop = np.array(A_flywheel + B_flywheel / r**2 - (1/8)*(1+3*v)*rho*(w**2)*r**2)
stress_axial = np.zeros(len(r))
stress = np.array([stress_radial, stress_hoop, stress_axial])
vm_stress = (1/sqrt(2)) * np.sqrt( (stress[0]-stress[1])**2 + (stress[1]-stress[2])**2 + (stress[2]-stress[0])**2 )

delta_r_flywheel = (Ri / E) * (stress_hoop[0] - v * (stress_radial[0] + stress_axial[0]))
print(f"radial change {delta_r_flywheel*1e6:.2f} microns")
```

```python
delta_r_change = delta_r_flywheel - delta_r_shaft
print(f"radial interference change {delta_r_change*1e6:.2f} microns")
delta_r_at_speed = delta_r - delta_r_change
print(f"radial interference at speed {delta_r_at_speed*1e6:.2f} microns")
pressure_at_speed = (E*delta_r_at_speed / (2*Ri)) * (1 - Ri**2 / Ro**2)
print(f"Interface pressure at speed {pressure_at_speed*1e-6:.2f} MPa")
```

# Stress in flywheel at speed

```python
B_flywheel = (-pressure_at_speed- (1/8)*(3+v)*rho*(w**2)*(Ro**2-Ri**2)) / ( (1/Ro**2) - (1/Ri**2) )
A_flywheel = B_flywheel / Ro**2 + (1/8)*(3+v)*rho*(w**2)*(Ro**2)
print(f"A = {A_flywheel:.2f}, B = {B_flywheel:.2f}")

r = np.linspace(Ri,Ro,100)
stress_radial = np.array(A_flywheel - B_flywheel / r**2 - (1/8)*(3+v)*rho*(w**2)*r**2)
stress_hoop = np.array(A_flywheel + B_flywheel / r**2 - (1/8)*(1+3*v)*rho*(w**2)*r**2)
stress_axial = np.zeros(len(r))
stress = np.array([stress_radial, stress_hoop, stress_axial])
vm_stress = (1/sqrt(2)) * np.sqrt( (stress[0]-stress[1])**2 + (stress[1]-stress[2])**2 + (stress[2]-stress[0])**2 )
print(f"Von mises stress inner {vm_stress[0]/1e6:.2f}MPa")
print(f"Radial stress inner {stress_radial[0]/1e6:.2f}MPa")
print(f"Hoop stress inner {stress_hoop[0]/1e6:.2f}MPa")
```

```python
fig, ax = plt.subplots()
ax.plot(r*1e3,1e-6*stress_radial, label = 'radial')
ax.plot(r*1e3,1e-6*stress_hoop, label = 'hoop')
ax.plot(r*1e3,1e-6*vm_stress, label = 'VM')
ax.set_xlabel("Radius (mm)")
ax.set_ylabel("Stress (MPa)")
ax.grid()
ax.legend()
```
